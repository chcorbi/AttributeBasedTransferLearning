```python
import numpy as np
import matplotlib.pyplot as plt
from utils import bzPickle, bzUnpickle
from sklearn.manifold import TSNE
from utils import get_class_attributes

p_type = "crosscorr"

classes = []
with open('./testclasses.txt') as infile:
    for line in infile:
        classes.append(line[:-1])

embeddings = get_class_attributes('./', 'test', p_type)
tsn = TSNE(perplexity=30, random_state=42)
embed_tsne = tsn.fit_transform(embeddings)

plt.figure(figsize=(10,9))
plt.xlim([-0.0002, 0.0002])
plt.ylim([-0.0002, 0.0002])
plt.scatter(embed_tsne[:,0], embed_tsne[:,1])

for i, txt in enumerate(classes):
    plt.annotate(txt, (embed_tsne[i,0],embed_tsne[i,1]))
plt.savefig('results/Embedding_'+p_type+'.jpg')
```
